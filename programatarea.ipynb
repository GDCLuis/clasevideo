```python
class Tarea:
    def __init__(self, nombre, descripcion, prioridad):
        self.nombre = nombre
        self.descripcion = descripcion
        self.prioridad = prioridad

    def __str__(self):
        return f"Tarea: {self.nombre}\nDescripción: {self.descripcion}\nPrioridad: {self.prioridad}"


class TareaGeneral(Tarea):
    def __init__(self, nombre, descripcion, prioridad):
        super().__init__(nombre, descripcion, prioridad)


class TareaConFecha(Tarea):
    def __init__(self, nombre, descripcion, prioridad, fecha):
        super().__init__(nombre, descripcion, prioridad)
        self.fecha = fecha

    def __str__(self):
        return super().__str__() + f"\nFecha: {self.fecha}"


class ListaDeTareas:
    def __init__(self):
        self.tareas = []

    def agregar_tarea(self, tarea):
        self.tareas.append(tarea)

    def eliminar_tarea(self, nombre):
        for tarea in self.tareas:
            if tarea.nombre == nombre:
                self.tareas.remove(tarea)
                return f"Tarea '{nombre}' eliminada."
        return f"Tarea '{nombre}' no encontrada."

    def listar_tareas(self):
        if not self.tareas:
            return "No hay tareas en la lista."
        return "\n\n".join(str(tarea) for tarea in self.tareas)


# Ejemplo de uso
if __name__ == "__main__":
    lista_de_tareas = ListaDeTareas()

    tarea1 = TareaGeneral("Comprar leche", "Comprar 2 litros de leche", "alta")
    tarea2 = TareaConFecha("Pagar factura", "Pagar la factura de electricidad", "media", "2024-06-15")

    lista_de_tareas.agregar_tarea(tarea1)
    lista_de_tareas.agregar_tarea(tarea2)

    print("Tareas en la lista:")
    print(lista_de_tareas.listar_tareas())

    print("\nEliminar una tarea:")
    print(lista_de_tareas.eliminar_tarea("Comprar leche"))

    print("\nTareas en la lista después de eliminar:")
    print(lista_de_tareas.listar_tareas())
```
